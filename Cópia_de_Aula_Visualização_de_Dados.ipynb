Desafio 1.1: Comparação de Vendas por Região

Cenário: Uma empresa de varejo quer comparar o desempenho de vendas de suas quatro regiões no último trimestre para identificar qual delas precisa de mais atenção.



```python
# Dados para o Desafio 1.1
dados_regiao = {
    'Regiao': ['Norte', 'Sul', 'Leste', 'Oeste'],
    'Vendas_Milhoes': [12.5, 18.2, 15.0, 9.8]
}
df_regiao = pd.DataFrame(dados_regiao)
print("\nDataFrame para Desafio 1.1:")
print(df_regiao)
```

1. Tipo de gráfico escolhido: Gráfico de barras verticais
2. Justificativa: Como existem 4 regiões e queremos comparar os valores, o gráfico de barras é o melhor para isso. De acordo o princípio de Cleveland-McGill comparar comprimento de barras em eixo comum é mais fácil para o as pessoas. De acordo com o princípio de Gestalt, barras uma do lado da outra melhoram a comparação.

```python
dados_regiao = {
    'Regiao': ['Norte', 'Sul', 'Leste', 'Oeste'],
    'Vendas_Milhoes': [12.5, 18.2, 15.0, 9.8]
}
df_regiao = pd.DataFrame(dados_regiao)

plt.figure(figsize=(8, 5))
sns.barplot(data=df_regiao, x='Regiao', y='Vendas_Milhoes', color='steelblue')
plt.title('Vendas por Região - Último Trimestre')
plt.xlabel('Região')
plt.ylabel('Vendas (em milhões R$)')
sns.despine()
plt.show()
```

Desafio 1.2: Distribuição de Idades de Clientes

Cenário: Uma equipe de marketing quer entender a distribuição de idade de seus clientes para segmentar campanhas. Eles têm dados de idade de 1000 clientes.

```python
# Dados para o Desafio 1.2
np.random.seed(42)
idades = np.random.normal(loc=35, scale=10, size=1000).astype(int)
idades = idades[(idades >= 18) & (idades <= 70)] # Limitar idades razoáveis
df_idades = pd.DataFrame({'Idade': idades})
print("\nDataFrame para Desafio 1.2 (primeiras 5 linhas):")
print(df_idades.head())
```

1. Tipo de gráfico escolhido: Histograma
2. Justificativa: Para ver como os dados estão distribuídos, o histograma é o mais apropiado. Ele junta as idades em faixas e mostra a frequência de cada uma. Pelo princípio de Gestalt as barras contínuas passam a ideia de continuidade, como a idade é uma variável contínua é o que faz mais sentido.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df_idades, x='Idade', bins=20, color='steelblue')
plt.title('Distribuição de Idades dos Clientes')
plt.xlabel('Idade')
plt.ylabel('Quantidade de Clientes')
sns.despine()
plt.show()
```

Desafio 1.3: Tendência de Acessos ao Site

Cenário: A equipe de produto quer monitorar a evolução diária do número de acessos ao site nos últimos 30 dias para identificar padrões ou quedas.

```python
# Dados para o Desafio 1.3
datas = pd.date_range(start='2024-03-01', periods=30)
np.random.seed(42)
acessos = np.random.randint(1000, 5000, 30)
# Simular uma queda nos últimos dias
acessos[-5:] = np.random.randint(500, 1500, 5)
df_acessos = pd.DataFrame({'Data': datas, 'Acessos': acessos})
print("\nDataFrame para Desafio 1.3 (primeiras 5 linhas):")
print(df_acessos.head())
```

1. Tipo de gráfico escolhido: Gráfico de linha
2. Justificativa: Como os dados são uma sequência de dias, o melhor é o gráfico de linha. De acordo com o princípio de Gestalt da continuidade, a linha conecta os pontos e deixa fácil enxergar a tendência ao longo do tempo.

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_acessos, x='Data', y='Acessos', color='steelblue')
plt.title('Acessos ao Site - Últimos 30 Dias')
plt.xlabel('Data')
plt.ylabel('Número de Acessos')
plt.xticks(rotation=45)
sns.despine()
plt.tight_layout()
plt.show()
```

Desafio 1.4: Relação entre Investimento em Marketing e Vendas

Cenário: A equipe de finanças quer entender se há uma correlação entre o investimento mensal em marketing e o volume de vendas gerado no mesmo mês.

```python
# Dados para o Desafio 1.4
np.random.seed(42)
investimento_marketing = np.random.randint(10000, 50000, 12)
vendas_mensais = investimento_marketing * 5 + np.random.randint(-5000, 5000, 12)
df_marketing_vendas = pd.DataFrame({
    'Investimento_Marketing': investimento_marketing,
    'Vendas_Mensais': vendas_mensais
})
print("\nDataFrame para Desafio 1.4 (primeiras 5 linhas):")
print(df_marketing_vendas.head())
```

1. Tipo de gráfico escolhido: Gráfico de dispersão
2. Justificativa: Já que queremos ver a relação entre duas variáveis numéricas contínuas. O gráfico de dispersão é o melhor pra isso. Pelo princípio de Cleveland-McGill conseguimos julgar a posição de cada ponto nos dois eixos ao mesmo tempo deixando a correlação entre investimento e vendas visível.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_marketing_vendas, x='Investimento_Marketing', y='Vendas_Mensais', color='steelblue')
plt.title('Relação entre Investimento em Marketing e Vendas')
plt.xlabel('Investimento em Marketing (R$)')
plt.ylabel('Vendas Mensais (R$)')
sns.despine()
plt.show()
```

Desafio 1.5: Composição de Mercado por Categoria de Produto

Cenário: Um analista de mercado quer visualizar a proporção de cada categoria de produto no total de vendas da empresa para entender a dominância de certas categorias.

Desafio 2.1: Vendas Trimestrais com Eixo Y Enganoso

Cenário: Um gerente de vendas recebeu este gráfico e está preocupado com uma suposta queda drástica nas vendas do Q4.

```python
# Dados para o Desafio 2.1
dados_trimestrais = {
    'Trimestre': ['Q1', 'Q2', 'Q3', 'Q4'],
    'Vendas': [1000, 1050, 1020, 980]
}
df_trimestral = pd.DataFrame(dados_trimestrais)

plt.figure(figsize=(7, 5))
ax = sns.barplot(x='Trimestre', y='Vendas', data=df_trimestral, color='lightcoral')
ax.set_ylim(bottom=950) # Eixo Y começando alto
ax.set_title('Vendas Trimestrais - Último Ano', fontsize=14, fontweight='bold')
ax.set_xlabel('Trimestre')
ax.set_ylabel('Vendas (Unidades)')
ax.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### Problemas identificados:

    a) Tufte - Lie Factor: o eixo Y começa em 950 ao invés de 0, exagerando visualmente as diferenças entre os trimestres. A queda do Q3 pro Q4 parece enorme, mas na prática é de apenas 40 unidades num total de ~1000.

    b) Tufte - Chartjunk: o grid tracejado não agrega informação relevante, é só ruído visual.

    c) Gestalt - Figura-fundo: a cor lightcoral chama atenção sem motivo, passando a impressão de que algo está errado quando os dados são quase estáveis.

```python
plt.figure(figsize=(7, 5))
ax = sns.barplot(x='Trimestre', y='Vendas', data=df_trimestral, color='steelblue')
ax.set_ylim(bottom=0)  # eixo começando do zero — honesto
ax.set_title('Vendas Trimestrais - Último Ano')
ax.set_xlabel('Trimestre')
ax.set_ylabel('Vendas (Unidades)')
sns.despine()
plt.show()
```

Desafio 2.2: Comparação de Produtos com Gráfico de Pizza Lotado

Cenário: A equipe de produto quer ver a proporção de vendas de 8 produtos diferentes, mas o gráfico atual está confuso.

```python
# Dados para o Desafio 2.2
dados_produtos = {
    'Produto': [f'Produto {i}' for i in range(1, 9)],
    'Vendas': [150, 120, 80, 70, 60, 50, 40, 30]
}
df_produtos = pd.DataFrame(dados_produtos)

plt.figure(figsize=(8, 8))
plt.pie(df_produtos['Vendas'], labels=df_produtos['Produto'], autopct='%1.1f%%', startangle=90)
plt.title('Proporção de Vendas por Produto', fontsize=14, fontweight='bold')
plt.show()
```

### Problemas identificados:

    a) Cleveland-McGill: pizza com 8 fatias é difícil de comparar. O olho humano não consegue julgar bem ângulos e áreas, especialmente nas fatias menores (Produto 5, 6, 7, 8) que ficam quase iguais visualmente.

    b) Tufte - Chartjunk: com 8 fatias, os rótulos e percentuais se acumulam e poluem o gráfico, tornando a leitura confusa.

    c) Gestalt - Semelhança: muitas fatias com tamanhos parecidos sem nenhuma distinção clara, o gráfico não comunica nada de imediato.

```python
df_produtos_ord = df_produtos.sort_values('Vendas', ascending=True)

plt.figure(figsize=(8, 5))
sns.barplot(data=df_produtos_ord, y='Produto', x='Vendas', color='steelblue')
plt.title('Proporção de Vendas por Produto')
plt.xlabel('Vendas (Unidades)')
plt.ylabel('Produto')
sns.despine()
plt.show()
```

Desafio 2.3: Desempenho de Funcionários com Cores Aleatórias

Cenário: O RH quer visualizar o desempenho (pontuação de 0 a 100) de 5 funcionários em uma avaliação, mas o gráfico atual usa cores sem sentido.

```python
# Dados para o Desafio 2.3
dados_funcionarios = {
    'Funcionario': ['Ana', 'Bruno', 'Carla', 'Daniel', 'Eva'],
    'Pontuacao': [85, 72, 91, 68, 79]
}
df_funcionarios = pd.DataFrame(dados_funcionarios)

plt.figure(figsize=(7, 5))
cores_aleatorias = ['red', 'green', 'blue', 'purple', 'orange']
ax = sns.barplot(x='Funcionario', y='Pontuacao', data=df_funcionarios, palette=cores_aleatorias)
ax.set_title('Pontuação de Desempenho dos Funcionários', fontsize=14, fontweight='bold')
ax.set_ylim(0, 100)
plt.show()
```

### Problemas identificados:
    a) Gestalt - Semelhança: cores diferentes em cada barra passam a impressão de que cada funcionário pertence a uma categoria diferente, quando na verdade são todos do mesmo grupo sendo comparados pela mesma métrica.
    b) Tufte - Chartjunk: as cores aleatórias são decoração pura, não codificam nenhuma informação útil. Só poluem o gráfico.
    c) Cleveland-McGill: sem ordenação, fica mais difícil identificar quem se saiu melhor ou pior de forma rápida.

```python
df_func_ord = df_funcionarios.sort_values('Pontuacao', ascending=False)

plt.figure(figsize=(7, 5))
ax = sns.barplot(x='Funcionario', y='Pontuacao', data=df_func_ord, color='steelblue')
ax.set_title('Pontuação de Desempenho dos Funcionários')
ax.set_xlabel('Funcionário')
ax.set_ylabel('Pontuação')
ax.set_ylim(0, 100)
sns.despine()
plt.show()
```

Desafio 2.4: Gráfico de Linhas com Chartjunk Excessivo

Cenário: Um analista de dados criou este gráfico para mostrar a evolução de duas métricas ao longo do tempo, mas ele está muito poluído.

```python
# Dados para o Desafio 2.4
datas_mes = pd.date_range(start='2023-01-01', periods=12, freq='M')
np.random.seed(42)
metrica_a = np.random.randint(50, 150, 12).cumsum()
metrica_b = np.random.randint(40, 130, 12).cumsum()

df_metricas = pd.DataFrame({
    'Data': datas_mes,
    'Metrica_A': metrica_a,
    'Metrica_B': metrica_b
})

plt.figure(figsize=(10, 6))
plt.plot(df_metricas['Data'], df_metricas['Metrica_A'], label='Métrica A', marker='o', linestyle='-', linewidth=2, color='blue')
plt.plot(df_metricas['Data'], df_metricas['Metrica_B'], label='Métrica B', marker='x', linestyle='--', linewidth=2, color='red')

plt.title('Evolução das Métricas A e B', fontsize=16, fontweight='bold')
plt.xlabel('Data', fontsize=12)
plt.ylabel('Valor', fontsize=12)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True, linestyle=':', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Problemas identificados:
    a) Tufte - Chartjunk: o grid com linhas pontilhadas em todo o fundo é desnecessário, só adiciona ruído visual sem ajudar a leitura dos dados.
    b) Tufte - Chartjunk: os marcadores em cada ponto (marker='o' e marker='x') também são excessivos numa série temporal contínua, poluem sem agregar informação.
    c) Gestalt - Proximidade: a legenda jogada pra fora do gráfico (bbox_to_anchor=(1,1)) fica distante das linhas, dificultando associar qual linha é qual métrica.

```python
plt.figure(figsize=(10, 5))
plt.plot(df_metricas['Data'], df_metricas['Metrica_A'], label='Métrica A', color='steelblue')
plt.plot(df_metricas['Data'], df_metricas['Metrica_B'], label='Métrica B', color='coral')
plt.title('Evolução das Métricas A e B')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
sns.despine()
plt.tight_layout()
plt.show()
```

Desafio 2.5: Comparação de Desempenho com Gráfico de Bolhas Confuso

Cenário: Um gerente de produto quer comparar o desempenho de diferentes funcionalidades (Funcionalidade 1 a 5) com base em duas métricas: 'Uso' e 'Satisfação'. O tamanho da bolha representa o 'Impacto'.

```python
# Dados para o Desafio 2.5
dados_funcionalidades = {
    'Funcionalidade': [f'Func {i}' for i in range(1, 6)],
    'Uso': [0.8, 0.6, 0.9, 0.4, 0.7],
    'Satisfacao': [0.9, 0.7, 0.8, 0.5, 0.6],
    'Impacto': [100, 50, 120, 30, 80]
}
df_funcionalidades = pd.DataFrame(dados_funcionalidades)

plt.figure(figsize=(8, 8))
plt.scatter(x=df_funcionalidades['Uso'], y=df_funcionalidades['Satisfacao'],
            s=df_funcionalidades['Impacto']*5, # Multiplicar para bolhas maiores
            alpha=0.7,
            c=np.random.rand(len(df_funcionalidades), 3)) # Cores aleatórias

for i, row in df_funcionalidades.iterrows():
    plt.text(row['Uso'], row['Satisfacao'], row['Funcionalidade'], fontsize=9, ha='center', va='center')

plt.title('Desempenho das Funcionalidades', fontsize=14, fontweight='bold')
plt.xlabel('Uso (0-1)', fontsize=12)
plt.ylabel('Satisfação (0-1)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)
plt.show()
```

### Problemas identificados:
    a) Gestalt - Semelhança: cores aleatórias em cada bolha não codificam nenhuma informação, só confundem. O olho tenta encontrar um padrão nas cores que simplesmente não existe.
    b) Tufte - Chartjunk: o grid pontilhado de fundo é desnecessário e compete visualmente com as bolhas.
    c) Cleveland-McGill: o tamanho das bolhas (área) é uma das codificações menos precisas para o olho humano. Sem legenda explicando o que o tamanho representa, o leitor nem sabe que o Impacto está sendo codificado ali.

```python
plt.figure(figsize=(8, 6))
plt.scatter(
    x=df_funcionalidades['Uso'],
    y=df_funcionalidades['Satisfacao'],
    s=df_funcionalidades['Impacto'] * 5,
    color='steelblue',
    alpha=0.6
)

for i, row in df_funcionalidades.iterrows():
    plt.text(row['Uso'], row['Satisfacao'] + 0.03, row['Funcionalidade'], fontsize=9, ha='center')

plt.title('Desempenho das Funcionalidades')
plt.xlabel('Uso (0-1)')
plt.ylabel('Satisfação (0-1)')
plt.figtext(0.5, -0.02, 'Tamanho da bolha = Impacto', ha='center', fontsize=9, color='gray')
sns.despine()
plt.tight_layout()
plt.show()
```
